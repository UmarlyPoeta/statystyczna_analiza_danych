## Zadanie 4

Napisz funkcję `sprawdz_bilety(wiek_pasazerow)`.

Funkcja przyjmuje listę lat osób w grupie: `[5, 10, 17, 22, 34, 6]`

**Działanie:**
Funkcja powinna przejść pętlą przez całą listę i dla każdej osoby sprawdzić warunek:
- Jeśli osoba ma mniej niż 7 lat - bilet jest darmowy (0 zł).
- Jeśli osoba ma od 7 do 18 lat - bilet kosztuje 15 zł.
- Jeśli osoba ma 18 lub więcej lat - bilet kosztuje 25 zł.
Funkcja ma wyświetlić komunikat dla każdej osoby (np. `"Wiek 10: 15 zł"`) i na końcu zwrócić sumaryczny koszt dla całej grupy.

```python
def sprawdz_bilety(wiek_pasazerow: list[int]):
  suma = 0
  for wiek in wiek_pasazerow:
    if wiek < 7:
      print(f"Wiek {wiek}: 0 zł")
    elif wiek < 18:
      print(f"Wiek {wiek}: 15 zł")
      suma += 15
    else:
      print(f"Wiek {wiek}: 25 zł")
      suma += 25
  return suma
```

## Zadanie 5
Napisz funkcję `bezpieczna_suma(liczby, limit=100)`.

1. Funkcja przechodzi pętlą przez listę liczby.
2. Dodaje do siebie kolejne liczby, ale tylko te, które są dodatnie.
3. Jeśli w trakcie dodawania suma przekroczy limit, funkcja przerywa pętlę i zwraca obecną sumę.

Aby sprawdzić działanie funkcji, przetestuj ją dla różnych list, np.:
- `liczby1 = [10, 20, 30, 40]`
- `liczby2 = [5, -3, 15, 50, 60]`
- `liczby3 = [1, 2, 3, 4, 5, 200]`

Sprawdź również działanie funkcji dla różnych wartości `limitu`.

```python
def bezpieczna_suma(liczby: list[int], limit: int = 100):
  suma = 0

  if limit < 0:
    return 0

  for liczba in liczby:
    if suma > limit:
      return suma
    if liczba < 0:
      continue
    suma += liczba
```

## Zadanie 6

Napisz funkcję `analizuj_wyniki(nazwa_kursu, *wyniki)`.

1. Funkcja przyjmuje najpierw nazwę przedmiotu, a potem dowolną liczbę punktów (0-100).
2. Najpierw musi przefiltrować dane: jeśli wynik jest spoza zakresu 0-100, funkcja ma go zignorować (użyj continue).
3. Następnie oblicza średnią z poprawnych wyników.
4. Na podstawie średniej przypisuje ocenę:
- `>= 90: "5"`
- `>= 70: "4"`
- `>= 50: "3"`
- `< 50: "2"`

Funkcja zwraca słownik: `{"kurs": nazwa_kursu, "srednia": obliczona_srednia, "ocena": wystawiona_ocena}`.

```python
def analizuj_wyniki(nazwa_kursu: str, *wyniki: int):
  wyniki_poprawne = []

  for wynik in wyniki:
    if wynik < 0 or wynik > 100:
      continue
    wyniki_poprawne.append(wynik)

  srednia = sum(wyniki_poprawne) / len(wyniki_poprawne)

  if srednia < 50:
    ocena = "2"
  elif srednia < 70:
    ocena = "3"
  elif srednia < 90:
    ocena = "4"
  else:
    ocena = "5"

  return {"kurs": nazwa_kursu, "srednia" : srednia, "ocena": ocena}
```
