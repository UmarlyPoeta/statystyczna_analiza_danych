# Zadanie 5
Utwórz tablicę 36 równomiernie rozmieszczonych punktów między 0 a 1, a następnie przekształć ją w macierz 6 × 6.

Następnie:
- Zwróć trzeci wiersz tej macierzy
- Zwróć ostatnią kolumnę
- Zwróć środkowy fragment macierzy 3 × 3





````python
punkty = np.linspace(0, 1, 36)
macierz = punkty.reshape(6, 6)

trzeci_wiersz = macierz[2, :]
ostatnia_kolumna = macierz[:, -1]
srodkowy_fragment_3x3 = macierz[1:4, 1:4]

macierz, trzeci_wiersz, ostatnia_kolumna, srodkowy_fragment_3x3
````

# Zadania 10
Stwórz poniższą macierz 4x4, a następnie:


$ \begin{bmatrix}
1 & 15 & 4 & 13 \\
8 & 21 & 3 & 12 \\
11 & 13 & 11 & 5 \\
32 & 13 & 0 & 2
\end{bmatrix}  $


- wyświetl element z drugiego wiersza i trzeciej kolumny,
- oblicz jej wyznacznik wykorzystując odpowiednią funkcję z biblioteki NumPy; **wskazówka:** użyj funkcji `np.linalg.det()`
- sumę elementów na głównej przekątnej,
- znajdź element największy i najmniejszy

````python
macierz_4x4 = np.array([
    [1, 15, 4, 13],
    [8, 21, 3, 12],
    [11, 13, 11, 5],
    [32, 13, 0, 2]
])

element_2_3 = macierz_4x4[1, 2]
wyznacznik = np.linalg.det(macierz_4x4)
suma_przekatnej = np.trace(macierz_4x4)
maksimum = np.max(macierz_4x4)
minimum = np.min(macierz_4x4)

element_2_3, wyznacznik, suma_przekatnej, maksimum, minimum
````

# Zadanie 13
Stwórz funkcję, która przyjmuje dwa wektory (oba o tym samym wymiarze) w postaci numpy tablicy, a zwróci odległość euklidesową między nimi.

 $ d\left( p,q\right)   = \sqrt {\sum _{i=1}^{n}  \left( q_{i}-p_{i}\right)^2 } $

````python
def fun(a, b):
    return np.sqrt(np.sum((b - a) ** 2))


fun(np.array([1, 2, 3]), np.array([4, 5, 6]))
````
